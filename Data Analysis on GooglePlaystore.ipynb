```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
df=pd.read_csv("googleplaystore.csv")
imputer=SimpleImputer(missing_values=np.nan, strategy="mean")
df.iloc[:,2:3]=imputer.fit_transform(df.iloc[:,2:3].values)
df=df.dropna()
```

```python
df.isna().sum()
```

# WITH MULTIPLE ROWS

## Q1. How many free apps are there in ART_AND_DESIGN?

```python
df=df.values
```

```python
df[0]
```

```python
cnt=0
for i in df:
    if(i[1]=="ART_AND_DESIGN" and i[6]=="Free"):
        cnt+=1
print("There are",cnt, "free apps in ART_AND_DESIGN")    
```

## Q2. How many apps in ART_AND_DESIGN have rating>4.5?

```python
cnt=0
for i in df:
    if(i[1]=="ART_AND_DESIGN" and i[2]>4.5):
        cnt+=1
print("There are",cnt, "apps with rating>4.5")    
```

## Q3. How many apps in FAMILY have rating>4.5 and FREE?

```python
cnt=0
for i in df:
    if(i[1]=="FAMILY" and i[2]>4.5 and i[6]=="Free"):
        cnt+=1
print("There are",cnt, "apps in FAMILY have rating>4.5 and FREE")    
```

## Q4. List all free apps in FAMILY with rating>4.5?

```python
print("-"*100)
print("These are all free apps in FAMILY with rating>4.5")
print("-"*100)

for i in df:
    if(i[1]=="FAMILY" and i[2]>4.5 and i[6]=="Free"):
        print(i[0])   
print("-"*100)
```

# USING groupby FROM PANDAS

## Q1. Name apps in ART_AND_DESIGN with rating>4.5 in ascending order WRT ratings

```python
df_pr=df[df["Category"]=="ART_AND_DESIGN"]
df_pr=df_pr[df_pr["Rating"]>4.5]
df_pr.sort_values(by ="Rating", ascending=True)
```

## Q2. Name apps in ART_AND_DESIGN in descending order WRT Ratings

```python
df_pr=df[df["Category"]=="ART_AND_DESIGN"]
df_pr.sort_values(by ="Rating", ascending=False)
```

## Q3. How many free apps are there in each category in Descending order

```python
df_pr=df[df["Type"]=="Free"]
df_pr.groupby("Category").count()["Type"].sort_values(ascending=False)
```
